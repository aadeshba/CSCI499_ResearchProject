```python
import json
from io import open

# Path to the .jsonl file
file_path = "2020_acl_diplomacy/data/train.jsonl"
diplomacy_data = {}

with open(file_path, 'r') as file:
    i = 0
    for line in file:
        # Parse the line as JSON
        diplomacy_data[i] = json.loads(line)
        i+=1


```
